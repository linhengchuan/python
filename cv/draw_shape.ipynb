```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int8)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, pt1=(50,50), pt2=(150,300), color=(255,0,0), thickness=10)
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img, center=(250,250), radius=50, color=(255,255,0), thickness=-1)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img, pt1=(100,100), pt2=(300,300), color=(255,100,100), thickness=10)
plt.imshow(blank_img)
```

```python
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(blank_img,text="Hello world!",org=(10,400),fontFace=font, fontScale=2, 
            color=(255,0,0), thickness=3, lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```
